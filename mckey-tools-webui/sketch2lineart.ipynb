Pythonのバージョンをテスト

```python
import sys
version = sys.version.split(' ')[0]
v_parts = version.split('.')
if int(v_parts[1]) < 10:
    print(f'Your Python version is less than 3.10 -> {version}')
else:
    print('Your Python version is good:', version)
```
